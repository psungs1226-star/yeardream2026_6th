# 지역변수 vs 전역변수
- 지역변수(Local) : 함수 안에서 정의 -> 함수 안에서만 존재(밖에서 접근 불가)
- 전역변수(Global) : 함수 밖에서 정의 -> 내부 프로그램 어디에서나 접근 가능

## 지역변수
- 함수 안에서만 존재

```python
def greet():
    msg = "함수 안에서 안녕!" # 지역변수
    print(msg)

greet()
```

```python
print(msg)
```

# 전역 변수
- 어디에서나 접근 가능

```python
msg = "Python is awesome!" # 전역변수

def display():
    print("함수 안:", msg)

display()
```

```python
print("함수 밖", msg)
```

## shadowing
- 같은 이름의 지역변수가 전역변수를 가림

```python
s = "I Love ueardreamschool"

def fun():
    s = "Me too."
    print(s)

fun()
```

```python
print(s)
```

# 전역변수를 수정하고 싶다면?
- global이 필요 => 아래에서 에러가 나온 이유는 로컬에서 전역변수를 수정하려 했기 때문에 발생

```python
count = 0
def increase():
    count += 1

increase()
```

```python
count = 0
def increase():
    global count # 이 선언이 있으면 전역을 바꿀 수 있음
    count += 1

increase()
```

```python
print(count)
```

```python
increase()
```

```python
print(count)
```

## 혼동개념
- 리턴을 해도 함수정의 내에서만 1로, 밖에서는 0으로 나옴

```python
count = 0
def increase(count=0):
    count += 1
    return count

increase()
```

```python
print(count)
```
